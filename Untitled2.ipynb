```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
data_train_path = r'C:\Users\GF63\Projek Certan\training\Chessman-image-dataset\Chess'
data_test_path = r'C:\Users\GF63\Projek Certan\test\Chessman-image-dataset\Chess'
data_val_path = r'C:\Users\GF63\Projek Certan\validation\Chessman-image-dataset\Chess'
```

```python
img_width, img_height = 180,180
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32,
    color_mode='grayscale' 
)
```

```python
data_chess = data_train.class_names
```

```python
data_chess
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(
    data_val_path,
    shuffle=False,
    image_size=(img_width, img_height),
    batch_size=32,
    color_mode='grayscale'  
)
```

```python

plt.figure(figsize=(10, 10))
for images, labels in data_train.take(1):
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().squeeze(), cmap='gray')  # Grayscale mode
        plt.title(data_chess[labels[i]])
        plt.axis('off')
```

```python
from tensorflow.keras.models import Sequential
```

```python
data_train
```

```python
model = Sequential([
    layers.Rescaling(1./255, input_shape=(img_width, img_height, 1)), 
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(data_chess)) 
])

```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
epochs_size = 20
history = model.fit(data_train, validation_data=data_val, epochs=epochs_size)
```

```python
epochs_range = range(epochs_size)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(epochs_range,history.history['accuracy'],label = 'Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'],label = 'Validation Accuracy')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range,history.history['loss'],label = 'Training Loss')
plt.plot(epochs_range, history.history['val_loss'],label = 'Validation Loss')
plt.title('Loss')
```

```python
image = r'C:\Users\GF63\Downloads\images.png'
image = tf.keras.utils.load_img(image, target_size=(img_height, img_width), color_mode='grayscale')
img_arr = tf.keras.utils.img_to_array(image)
img_bat = tf.expand_dims(img_arr, axis=0)
```

```python
predict = model.predict(img_bat)
```

```python
score = tf.nn.softmax(predict)
```

```python
print('Chess in image is {} with accuracy of {:0.2f}'.format(data_chess[np.argmax(score)],np.max(score)*100))
```
