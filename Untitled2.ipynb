```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras import Sequential, layers, Input
```

```python
data_train_path = r'C:\Users\GF63\Projek Certan\training\Chessman-image-dataset\Chess'
data_test_path = r'C:\Users\GF63\Projek Certan\test\Chessman-image-dataset\Chess'
data_val_path = r'C:\Users\GF63\Projek Certan\validation\Chessman-image-dataset\Chess'
```

```python
img_width, img_height = 180,180
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32,
    validation_split=False)
```

```python
data_chess = data_train.class_names
```

```python
data_chess
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(data_val_path,
                                                       image_size=(img_height,img_width),
                                                       batch_size=32,
                                                        shuffle=False,
                                                       validation_split=False)
```

```python
plt.figure(figsize=(10,10))
for image, labels in data_train.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_chess[labels[i]])
        plt.axis('off')
```

```python
from tensorflow.keras.models import Sequential
```

```python
data_train
```

```python
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1)
])

model = Sequential([
    Input(shape=(180, 180, 3)),
    # Data normalization
    layers.Rescaling(1./255),
    
    # First convolutional block
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    # Second convolutional block
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    # Third convolutional block
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    # Fourth convolutional block (added)
    layers.Conv2D(128, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    # Fifth convolutional block (added)
    layers.Conv2D(256, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    # Flatten the output
    layers.Flatten(),
    
    # Dropout layer to prevent overfitting
    layers.Dropout(0.2),
    
    # Fully connected layers
    layers.Dense(128, activation='relu'),
    layers.Dense(len(data_chess))  # Output layer with classes equal to `len(data_chess)`
])
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
model.summary()
```

```python
epochs_size = 20
history = model.fit(data_train, validation_data=data_val, epochs=epochs_size)
```

```python
epochs_range = range(epochs_size)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(epochs_range,history.history['accuracy'],label = 'Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'],label = 'Validation Accuracy')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range,history.history['loss'],label = 'Training Loss')
plt.plot(epochs_range, history.history['val_loss'],label = 'Validation Loss')
plt.title('Loss')
```
